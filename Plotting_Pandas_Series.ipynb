```python
import pandas as pd
from pandas import Series
from pandas import DataFrame
from pandas_datareader import data
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = data.DataReader("MSFT", "yahoo", start=dt.datetime(2017, 2, 1), end=dt.datetime(2018, 1, 25))
```

```python
close_px = df['Adj Close']
mavg = close_px.rolling(window=40, center=False).mean() # pd.rolling_mean(close_px, 40)
```

```python
rets = close_px/close_px.shift(1) - 1
close_px.pct_change().head()
close_px.plot(label='MSFT')
mavg.plot(label='Moving average', color='red')
plt.legend()
plt.show()
```
